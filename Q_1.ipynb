#Crammer Matrix

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Select categorical columns
categorical_cols = df[['Gender', 'Age', 'Ethnicity']]

# Compute Cramer's V for each pair of categorical columns
cramer_v_matrix = np.zeros((len(categorical_cols.columns), len(categorical_cols.columns)))
for i in range(len(categorical_cols.columns)):
    for j in range(len(categorical_cols.columns)):
        contingency_table = pd.crosstab(categorical_cols.iloc[:, i], categorical_cols.iloc[:, j])
        chi2, _, _, _ = chi2_contingency(contingency_table)
        n = contingency_table.sum().sum()
        cramer_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))
        cramer_v_matrix[i, j] = cramer_v

# Display Cramer's V matrix
cramer_v_df = pd.DataFrame(cramer_v_matrix, index=categorical_cols.columns, columns=categorical_cols.columns)
print("Cramer's V Matrix:")
print(cramer_v_df)
```

#Custom linear Regressor

```python
class CustomLinearRegressor:
    def __init__(self):
        self.coefficients = None
        self.intercept = None

    def fit(self, X, y):
        # Add a column of ones to X for the intercept term
        X_augmented = np.hstack((np.ones((X.shape[0], 1)), X))

        # Compute the coefficients using the normal equation (closed-form solution)
        self.coefficients = np.linalg.inv(X_augmented.T @ X_augmented) @ X_augmented.T @ y

        # Extract the intercept and coefficients
        self.intercept = self.coefficients[0]
        self.coefficients = self.coefficients[1:]

    def predict(self, X):
        # Add a column of ones to X for the intercept term
        X_augmented = np.hstack((np.ones((X.shape[0], 1)), X))

        # Predict the target variable
        y_pred = X_augmented @ np.concatenate(([self.intercept], self.coefficients))
        return y_pred
```

```python
import numpy as np

class CustomPseudoInverseRegressor:
    def __init__(self):
        self.coefficients = None

    def fit(self, X, y):
        # Add a column of ones to X for the intercept term
        X_augmented = np.hstack((np.ones((X.shape[0], 1)), X))

        # Compute the pseudo-inverse of X
        X_pseudo_inverse = np.linalg.pinv(X_augmented)

        # Compute the coefficients using the pseudo-inverse
        self.coefficients = X_pseudo_inverse @ y

    def predict(self, X):
        # Add a column of ones to X for the intercept term
        X_augmented = np.hstack((np.ones((X.shape[0], 1)), X))

        # Predict the target variable
        y_pred = X_augmented @ self.coefficients
        return y_pred
```

#custom regressor with gradient

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Define the CustomGradientLinearRegressor class
class CustomGradientLinearRegressor:
    def __init__(self, learning_rate=0.01, max_iter=1000, tol=1e-4):
        self.coefficients = None
        self.intercept = None
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tol = tol
        self.loss_history = []  # List to store loss at each iteration

    def fit(self, X, y):
        # Add a column of ones to X for the intercept term
        X_augmented = np.hstack((np.ones((X.shape[0], 1)), X))

        # Initialize coefficients randomly
        np.random.seed(42)  # For reproducibility
        self.coefficients = np.random.randn(X_augmented.shape[1])

        # Gradient descent optimization
        prev_loss = float('inf')
        for i in range(self.max_iter):
            # Compute predictions
            y_pred = X_augmented @ self.coefficients

            # Compute gradient of the loss with respect to coefficients
            gradient = 2 * X_augmented.T @ (y_pred - y) / X_augmented.shape[0]

            # Update coefficients using gradient descent
            self.coefficients -= self.learning_rate * gradient

            # Compute loss (mean squared error)
            loss = np.mean((y_pred - y) ** 2)

            # Store the loss in the history
            self.loss_history.append(loss)

            # Check convergence
            if abs(loss - prev_loss) < self.tol:
                print(f"Converged after {i} iterations")
                break

            prev_loss = loss

        # Extract the intercept and coefficients
        self.intercept = self.coefficients[0]
        self.coefficients = self.coefficients[1:]

    def predict(self, X):
        # Add a column of ones to X for the intercept term
        X_augmented = np.hstack((np.ones((X.shape[0], 1)), X))

        # Predict the target variable
        y_pred = X_augmented @ np.concatenate(([self.intercept], self.coefficients))
        return y_pred



```

#Noise Injection

```python
class CustomGradientLinearRegressor:
    def __init__(self, learning_rate=0.01, max_iter=1000, tol=1e-4, input_noise=0.0, param_noise=0.0):
        self.coefficients = None
        self.intercept = None
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tol = tol
        self.loss_history = []  # List to store loss at each iteration
        self.input_noise = input_noise  # Noise level to inject into inputs
        self.param_noise = param_noise  # Noise level to inject into parameters

    def fit(self, X, y):
        # Add random noise to the input features
        X_noisy = X + np.random.normal(scale=self.input_noise, size=X.shape)

        # Add a column of ones to X for the intercept term
        X_augmented = np.hstack((np.ones((X_noisy.shape[0], 1)), X_noisy))

        # Initialize coefficients randomly with noise
        np.random.seed(42)  # For reproducibility
        self.coefficients = np.random.randn(X_augmented.shape[1]) + np.random.normal(scale=self.param_noise, size=X_augmented.shape[1])

        # Gradient descent optimization with noise
        prev_loss = float('inf')
        for i in range(self.max_iter):
            # Compute predictions
            y_pred = X_augmented @ self.coefficients

            # Compute gradient of the loss with respect to coefficients with noise
            gradient = 2 * X_augmented.T @ (y_pred - y) / X_augmented.shape[0] + np.random.normal(scale=self.param_noise, size=self.coefficients.shape)

            # Update coefficients using gradient descent
            self.coefficients -= self.learning_rate * gradient

            # Compute loss (mean squared error)
            loss = np.mean((y_pred - y) ** 2)

            # Store the loss in the history
            self.loss_history.append(loss)

            # Check convergence
            if abs(loss - prev_loss) < self.tol:
                print(f"Converged after {i} iterations")
                break

            prev_loss = loss

        # Extract the intercept and coefficients
        self.intercept = self.coefficients[0]
        self.coefficients = self.coefficients[1:]
```
